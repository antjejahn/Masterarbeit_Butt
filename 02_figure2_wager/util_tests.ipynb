### Tests für inf_JK_bagged_variance Funktion

```python
import unittest
import numpy as np
from utils import inf_JK_bagged_variance


class TestInfJKBaggedVariance(unittest.TestCase):
    def setUp(self):
        self.N_bi = np.array([[2, 0, 1], [0, 0, 3], [1, 0, 2]])
        self.T_N_b = np.array([[6, 8], [10, 15], [5, 7]])
        self.weights2 = np.array([0.7, 0, 0.3])

    def test_without_weights(self):
        res1, res2 = inf_JK_bagged_variance(self.N_bi, self.T_N_b)
        expected_result = np.array([4/9, 22/9])
        np.testing.assert_array_equal(np.round(res1 - res2,13), np.round(expected_result,13))

# To run the tests in a Jupyter notebook
def run_tests():
    loader = unittest.TestLoader()
    suite = loader.loadTestsFromTestCase(TestInfJKBaggedVariance)
    runner = unittest.TextTestRunner(verbosity=2)
    runner.run(suite)


# Run the tests
run_tests()
```
