```python
from PIL import Image

# Pfade zu den Bildern
a_path = r"C:\Users\rehan\meine_Repos\Masterarbeit\Paper\help_figures\high_S\figure3_morris_high_S.png"
b_path = r"C:\Users\rehan\meine_Repos\Masterarbeit\Paper\help_figures\high_S\figure4_morris_high_S.png"
c_path = r"C:\Users\rehan\meine_Repos\Masterarbeit\Paper\help_figures\high_S\pred_S_bias(n_train)1000__(B_RF)1000__(n_sim)1000__covariates4__10.png"
d_path = r"C:\Users\rehan\meine_Repos\Masterarbeit\Paper\help_figures\high_S\pred_var_bias(n_train)1000__(B_RF)1000__(n_sim)1000__covariates4__350.png"

# Bilder öffnen
a = Image.open(a_path)
b = Image.open(b_path)
c = Image.open(c_path)
d = Image.open(d_path)

# Größen der Bilder
a_width, a_height = a.size
b_width, b_height = b.size

# grösse von b ändern 
b = b.resize((a_width, a_height))
b_width, b_height = b.size

c_width, c_height = c.size
# grösse von c ändern

c = c.resize((a_width, int(a_height/2)))
c_width, c_height = c.size
d = d.resize((a_width, int(a_height/2)))
d_width, d_height = d.size

# Neue Bildgröße berechnen
new_width = a_width + b_width + max(c_width, d_width)
new_height = max(a_height, b_height, c_height + d_height)

# Neues Bild erstellen
new_image = Image.new('RGB', (new_width, new_height))

# Bilder einfügen
new_image.paste(a, (0, 0))
new_image.paste(b, (a_width, 0))
new_image.paste(c, (a_width + b_width, 0))
new_image.paste(d, (a_width + b_width, c_height))

# Neues Bild speichern
new_image.save(r'C:\Users\rehan\meine_Repos\Masterarbeit\Paper\help_figures\comb_figs_high_S.png')
```
