```python
import os
from PIL import Image

pp = r"C:\Users\rehan\OneDrive\Desktop\Neuer Ordner"

# Lade alle PNG-Bilder aus dem Ordner in eine Liste
images = [Image.open(os.path.join(pp, f)) for f in os.listdir(pp) if f.endswith('.png')]

# Stelle sicher, dass du genau 4 Bilder hast
if len(images) != 4:
    raise ValueError("Es wurden nicht genau 4 PNG-Bilder im Ordner gefunden.")

# Ermittle die maximalen Breiten und Höhen der 4 Bilder
widths, heights = zip(*(img.size for img in images))
max_width = max(widths)
max_height = max(heights)

# Wir erstellen ein neues Bild mit 2 Spalten und 2 Zeilen
# => Gesamtbreite = 2 * maximale Breite
# => Gesamthöhe    = 2 * maximale Höhe
new_width = 2 * max_width
new_height = 2 * max_height

# Neues, leeres Bild anlegen
new_img = Image.new("RGB", (new_width, new_height))

# Bilder einfügen:
#   - Bild 0 links oben
#   - Bild 1 rechts oben
#   - Bild 2 links unten
#   - Bild 3 rechts unten
new_img.paste(images[2], (0, 0))
new_img.paste(images[3], (max_width, 0))
new_img.paste(images[0], (0, max_height))
new_img.paste(images[1], (max_width, max_height))

# Speichere das Ergebnis, z. B. unter dem Namen "collage.png"
output_path = os.path.join(pp, "collage.png")
new_img.save(output_path)
print(f"Collage wurde erfolgreich unter {output_path} gespeichert.")
```
