### Tests für inf_JK_bagged_variance Funktion

```python
import unittest
import numpy as np
from utils import inf_JK_bagged_variance


class TestInfJKBaggedVariance(unittest.TestCase):
    def setUp(self):
        self.N_bi = np.array([[2, 0], [0, 2], [1, 1]])
        self.N_bi_2 = np.array([[2, 0, 1], [0, 0, 3], [1, 0, 2]])
        self.T_N_b = np.array([[6, 8], [10, 15], [5, 7]])
        self.weights2 = np.array([0.7, 0, 0.3])

    def test_without_weights(self):
        res1, res2 = inf_JK_bagged_variance(self.N_bi, self.T_N_b)
        expected_result = np.array([2.75, 10.25])
        np.testing.assert_array_equal(res1 - res2, expected_result)

    def test_with_weights(self):
        weights = np.array([1 / 3, 1 / 3, 1 / 3])
        res1, res2 = inf_JK_bagged_variance(self.N_bi, self.T_N_b, weights)
        expected_result = np.array([2.75, 10.25])
        np.testing.assert_array_equal(res1 - res2, expected_result)

    def test_with_zero_weights(self):
        weights = np.array([0.7, 0, 0.3])
        res1, res2 = inf_JK_bagged_variance(
            N_bi=self.N_bi_2, T_N_b=self.T_N_b, weights=weights
        )
        expected_result = np.array([(23641 / 3969) - (7 / 4), (1459 / 81) - (19 / 4)])
        np.testing.assert_array_equal(
            np.round(res1 - res2, 13), np.round(expected_result, 13)
        )


# To run the tests in a Jupyter notebook
def run_tests():
    loader = unittest.TestLoader()
    suite = loader.loadTestsFromTestCase(TestInfJKBaggedVariance)
    runner = unittest.TextTestRunner(verbosity=2)
    runner.run(suite)


# Run the tests
run_tests()
```
